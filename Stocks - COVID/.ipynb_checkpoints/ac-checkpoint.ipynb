```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime as dt
```

```python
AC.dtypes
```

```python
AC.Date = pd.to_datetime(AC.Date)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))


# Plot a line and add some simple annotations
line, = ax.plot(AC.Date, AC.High, linewidth=3)

'''
ax.annotate('First Covid Death in Ontario',
            xy=(dt.date(2020,3,11),30.4 ), xycoords='data',
            xytext=(0.3, 0.95), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right', verticalalignment='top')
'''


ax.set_xticklabels([])
ax.set(ylim=(10, 55))
plt.vlines(dt.date(2020,3,11),10,55, colors='k', linestyles='solid', label='First COVID Death in Ontario', linewidth=0.5)
plt.vlines(dt.date(2020,1,25),10,55, colors='r', linestyles='solid', label='First Presumptive Case', linewidth=0.5)
plt.vlines(dt.date(2020,3,17),10,55, colors='k', linestyles='solid', label='State of Emergency in Ontario', linewidth=0.5)
plt.vlines(dt.date(2020,3,23),10,55, colors='k', linestyles='solid', label='Non Essential Businesses closed in Ontario', linewidth=0.5)

plt.show()
```

```python
date_df = pd.read_excel('Covid-Dates.xlsx')
date_df = pd.DataFrame(date_df)
date_df.dtypes
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

# Plot a line and add some simple annotations
line, = ax.plot(AC.Date, AC.High, linewidth=3)

#ax.set_xticklabels([])

plt.xticks(ticks=date_df['Date'], labels=date_df['Short Text'], rotation=90)

ax.set(ylim=(10, 55))

for row in date_df:
    plt.vlines(date_df.Date, 10,55,colors='k', linewidth=0.5)

plt.show()
```

```python
AC_2019 = pd.DataFrame(pd.read_csv('AC-2019.csv'))
AC_2019
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(20,15))

fig.subplots_adjust(hspace=0.5)
plt.title('Air Canada Stock Prices')

ax1.set_ylim(10, 55)
ax1.plot(AC_2019.Date, AC_2019.High)
ax1.set_xlabel('2019')
ax1.set_ylabel('Stock Price (CAD)')

ax2.set_ylim(10, 55)
ax2.plot(AC.Date, AC.High, color='r')
ax2.set_xlabel('2020')
ax2.set_ylabel('Stock Price (CAD)')

ax1.set_xticklabels([])
ax2.set_xticklabels([])

plt.tight_layout()
plt.show()
```
